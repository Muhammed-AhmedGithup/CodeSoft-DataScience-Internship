```python
from sklearn.preprocessing import StandardScaler
```

```python
df['Amount']=StandardScaler().fit_transform(df[['Amount']])
df['Time']=StandardScaler().fit_transform(df[['Time']])
```

```python
df=df.sample(frac=1)
f=df.loc[df['Class']==1]
non_f=df.loc[df['Class']==0][:492]
new_d=pd.concat([f,non_f])
df=new_d.sample(frac=1,random_state=42)
df.head()
```

```python
for col in df.columns:
    if col=='Class':
        continue
    else:
        plt.figure(figsize=(8,5))
        sns.boxplot(x='Class',y=col,data=df)
        plt.title('Class '+'and '+col)
        plt.show()
```

```python
for col in df.columns:
    if col=='Class':
        continue
    else:
        Q1=df[col].quantile(0.25)
        Q3=df[col].quantile(0.75)
        IQR=Q3-Q1
        l=Q1-1.5*IQR
        u=Q3+1.5*IQR
        df=df.drop(df[(df[col]>u)|(df[col]<l)].index)
```

```python
df.shape
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df.drop('Class',axis=1)
y=df['Class']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import roc_auc_score
```

```python
models={'LogisticRegression':LogisticRegression(),
         'DecisionTreeClassifier':DecisionTreeClassifier(),
         'KNeighborsClassifier':KNeighborsClassifier(),
         'SVC':SVC()}
```

```python
for name,model in models.items():
    print('the model is '+name)
    model.fit(x_train,y_train)
    yhat=model.predict(x_test)
    print('the accuracy: ',accuracy_score(y_test,yhat))
    print('-------------------------------------------------------')
```
